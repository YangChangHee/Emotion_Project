```python
def draw_2_image(image,image2):
    rows=1
    cols=2
    fig=plt.figure()
    ax1=fig.add_subplot(rows,cols,1)
    ax1.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    ax1.axis("off")

    ax2=fig.add_subplot(rows,cols,2)
    ax2.imshow(cv2.cvtColor(image2, cv2.COLOR_BGR2RGB))
    ax2.axis("off")
    plt.show()
```

```python
def Image_Result(image):
    img=image
    gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    for (x,y,w,h) in faces:
        img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = img[y:y+h, x:x+w]
    draw_2_image(img,roi_color)
```

```python
face_cascade = cv2.CascadeClassifier('/home/lab/양창희/Emotion Project/Haarcascades/haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('/home/lab/양창희/Emotion Project/Haarcascades/haarcascade_eye.xml')
eyes_cascade = cv2.CascadeClassifier('/home/lab/양창희/Emotion Project/Haarcascades/frontalEyes.xml')
nose_cascade = cv2.CascadeClassifier('/home/lab/양창희/Emotion Project/Haarcascades/Nariz.xml')
mouth_cascade = cv2.CascadeClassifier('/home/lab/양창희/Emotion Project/Haarcascades/Mouth.xml')
```

```python
def Image_Faces(image):
    img=cv2.imread(image)
    gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    for (x,y,w,h) in faces:
        img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = img[y:y+h, x:x+w]
    draw_2_image(img,roi_color)
```

```python
img='/home/lab/양창희/Emotion Project/Test_image/test_1.jpg'
Image_Faces(img)
```

```python
def image_eye(image):
    img=cv2.imread(image)
    gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    eye = eye_cascade.detectMultiScale(gray, 1.3, 5)
    for (x,y,w,h) in eye:
        img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = img[y:y+h, x:x+w]
    draw_2_image(img,roi_color)
```
