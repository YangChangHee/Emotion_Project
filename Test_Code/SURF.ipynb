# SURF

```python
import cv2
import matplotlib.pyplot as plt
img='/home/lab//Emotion Project/Test_image/test_1.jpg'
img=cv2.imread(img)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

# CLOCKWISE 90

```python
img90 = cv2.rotate(img,cv2.ROTATE_90_CLOCKWISE)
```

```python
import os
def draw_2_image(image,image2):
    rows=1
    cols=2
    fig=plt.figure()
    ax1=fig.add_subplot(rows,cols,1)
    ax1.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    ax1.axis("off")

    ax2=fig.add_subplot(rows,cols,2)
    ax2.imshow(cv2.cvtColor(image2, cv2.COLOR_BGR2RGB))
    ax2.axis("off")
    plt.show()
```

```python
def draw_2_gray_image(image,image2):
    rows=1
    cols=2
    fig=plt.figure()
    ax1=fig.add_subplot(rows,cols,1)
    ax1.imshow(image, cmap='gray')
    ax1.axis("off")

    ax2=fig.add_subplot(rows,cols,2)
    ax2.imshow(image2, cmap='gray')
    ax2.axis("off")
    plt.show()
```

```python
draw_2_image(img,img90)
```

# Gray Transform

```python
imgray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
imgray90=cv2.cvtColor(img90,cv2.COLOR_BGR2GRAY)
```

```python
draw_2_gray_image(imgray,imgray90)
```

# OpenCv SURF Function Load
## Threshold => 1000

```python
img2, img3 = None,None
surf = cv2.xfeatures2d.SURF_create()
surf.setHessianThreshold(1000)
```

# Keypoint Descriptor

```python
kp1, desc1 = surf.detectAndCompute(imgray, None)
kp2, desc2 = surf.detectAndCompute(imgray90, None)
```

```python
img2 = cv2.drawKeypoints(imgray, kp1, img2)
img3 = cv2.drawKeypoints(imgray90, kp2, img3)
```

# Keypoint Draw

```python
draw_2_image(img2,img3)
```

```python
img4, img5 = None,None
img4 = cv2.drawKeypoints(imgray, kp1, img4,
                        flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
img5 = cv2.drawKeypoints(imgray90, kp2, img5,
                        flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
```

# Keypoint and Flag Draw

```python
draw_2_image(img4,img5)
```

# Image Matching

```python
matcher = cv2.BFMatcher(cv2.NORM_L1, crossCheck=True)
```

```python
matches = matcher.match(desc1, desc2)
```

```python
res = cv2.drawMatches(img, kp1, img90, kp2, matches, None, \
                      flags=cv2.DRAW_MATCHES_FLAGS_NOT_DRAW_SINGLE_POINTS)
```

```python
plt.imshow(cv2.cvtColor(res, cv2.COLOR_BGR2RGB))
```

```python
def SUSRF_Draw_Image(image):
    img=image
    img90 = cv2.rotate(img,cv2.ROTATE_90_CLOCKWISE)
    draw_2_image(img,img90)
    
    imgray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    imgray90=cv2.cvtColor(img90,cv2.COLOR_BGR2GRAY)
    draw_2_gray_image(imgray,imgray90)
    
    img2, img3 = None,None
    surf = cv2.xfeatures2d.SURF_create()
    kp1, desc1 = surf.detectAndCompute(imgray, None)
    kp2, desc2 = surf.detectAndCompute(imgray90, None)
    img2 = cv2.drawKeypoints(imgray, kp1, img2)
    img3 = cv2.drawKeypoints(imgray90, kp2, img3)
    draw_2_image(img2,img3)
    
    img4, img5 = None,None
    img4 = cv2.drawKeypoints(imgray, kp1, img4,
                            flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
    img5 = cv2.drawKeypoints(imgray90, kp2, img5,
                            flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
    draw_2_image(img4,img5)
    
    matcher = cv2.BFMatcher(cv2.NORM_L1, crossCheck=True)
    matches = matcher.match(desc1, desc2)
    res = cv2.drawMatches(img, kp1, img90, kp2, matches, None, \
                      flags=cv2.DRAW_MATCHES_FLAGS_NOT_DRAW_SINGLE_POINTS)
    plt.imshow(cv2.cvtColor(res, cv2.COLOR_BGR2RGB))
```

```python
def SURF(image):
    img=image
    img90 = cv2.rotate(img,cv2.ROTATE_90_CLOCKWISE)
    
    imgray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    imgray90=cv2.cvtColor(img90,cv2.COLOR_BGR2GRAY)
    
    img2, img3 = None,None
    surf = cv2.xfeatures2d.SURF_create()
    kp1, desc1 = surf.detectAndCompute(imgray, None)
    kp2, desc2 = surf.detectAndCompute(imgray90, None)
    img2 = cv2.drawKeypoints(imgray, kp1, img2)
    img3 = cv2.drawKeypoints(imgray90, kp2, img3)
    
    img4, img5 = None,None
    img4 = cv2.drawKeypoints(imgray, kp1, img4,
                            flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
    img5 = cv2.drawKeypoints(imgray90, kp2, img5,
                            flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
    
    matcher = cv2.BFMatcher(cv2.NORM_L1, crossCheck=True)
    matches = matcher.match(desc1, desc2)
    res = cv2.drawMatches(img, kp1, img90, kp2, matches, None, \
                      flags=cv2.DRAW_MATCHES_FLAGS_NOT_DRAW_SINGLE_POINTS)
    plt.imshow(cv2.cvtColor(res, cv2.COLOR_BGR2RGB))
```
